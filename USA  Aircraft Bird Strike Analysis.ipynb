## Data Type Conversion

#### Converting 'Aircraft: Number of engines?' to integer form

```python
df['Aircraft: Number of engines?']=df['Aircraft: Number of engines?'].replace('C',np.nan)
    
df['Aircraft: Number of engines?'].unique()
```

```python
df['Aircraft: Number of engines?']=df['Aircraft: Number of engines?'].fillna(-1).astype(int)
df['Aircraft: Number of engines?']
```

#### Converting  'Cost: Total $' &  'Feet above ground' to integer forms

```python


obj_int=['Cost: Total $','Feet above ground']

for i in obj_int:
    
    df[i] = df[i].replace(',', '', regex=True).fillna(-1).astype(int)
    
    
  
 #converting 'FlightDate' from object to date format

df['FlightDate'] = pd.to_datetime(df['FlightDate'], errors='coerce')
print(df[['Cost: Total $','Feet above ground']])
df.head(2)
```

#### Defining numerical and categorical variables

```python
numerical=['Wildlife: Number Struck Actual', 'Aircraft: Number of engines?', 'Cost: Total $', 
           'Feet above ground','Number of people injured']



categorical=['Airport: Name', 'Altitude bin',
       'Aircraft: Make/Model', 'Wildlife: Number struck',
       'FlightDate', 'Effect: Indicated Damage',
      'Aircraft: Airline/Operator',
       'Origin State', 'When: Phase of flight',
       'Remains of wildlife collected?',
       'Remains of wildlife sent to Smithsonian', 'Wildlife: Size',
       'Conditions: Sky', 'Wildlife: Species',
       'Pilot warned of birds or wildlife?', 
       'Is Aircraft Large?']



for i in numerical:
    print(f"\n {i}: \n", df[i].describe())
          
          
for i in categorical:
          print(f"\n {i}: \n", df[i].mode())
```

```python
# computing skewness 

for i in [numerical[1],numerical[3]]:
    print(f"Skewness of {i}: {df[i].skew()}")
```

Since, both have skewed distribution, median will be used to impute missing values.

## Handling Missing Values

```python
for i in [numerical[1], numerical[3]] :
    df[i]= df[i].replace(-1, df[i].median())
```

```python
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

### Visualising Variables(Columns)

```python

n_cols = 2
n_rows = (len(m) + n_cols - 1) // n_cols  # Calculate the number of rows needed
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 6 * n_rows))

# Flattening axes array for easy iteration
axes = axes.flatten()

# Plotting countplots for each column
for i, col in enumerate(m):
    sns.countplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].set_ylabel('Frequency')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
